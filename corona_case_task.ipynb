Task1 plot subplot2grid with the datas

 Import librabries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Reading files

```python
i=pd.read_excel("/content/coronacases.xls")
i
```

```python
df=pd.DataFrame(i)
df
```

```python
df.describe()
```

```python
df.info
```

```python
e=df['Unnamed: 0']
e
```

```python
l=df['cases']
l
```

```python
v=df['active cases']
v
```

```python
p=df['deaths']
p
```

```python
from matplotlib.pyplot import figure 
figure(figsize=(100,100),dpi=100)
fig=plt.figure()
s1=plt.subplot2grid((3,5),(0,1),colspan=3)
s2=plt.subplot2grid((3,5),(1,0),colspan=2,rowspan=3)
s3=plt.subplot2grid((3,5),(1,3),colspan=5,rowspan=3)
s1.plot(e,l,label="corona total no of cases",color="g")
plt.xlabel("date")
plt.ylabel("total no of cases")
plt.xticks(rotation=90)
s2.scatter(e,v,label="corona active cases",color="r")
plt.xlabel("date")
plt.ylabel("active cases")
plt.xticks(rotation=90)
s3.bar(e,p,label="corona death cases",color="c")
plt.xlabel("date")
plt.ylabel("death cases")
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
h = h.dropna()
h = h.reset_index()
h
```

```python
Ind= h[h['country'] == 'India']
Ind
```

```python
Ind.head()
```

```python
o=Ind['daily_vaccinations'].value_counts()
o
```

```python
from matplotlib.pyplot import figure 
figure(figsize=(50,8), dpi=100)
x=Ind['date']
y=Ind['daily_vaccinations']
plt.bar(x,y,label='daily_vaccinations',color="b")
plt.xlabel("date")
plt.ylabel("daily_vaccinations")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
Aus= h[h['country'] == 'Australia']
Aus.describe()
```

```python
total = h.groupby('country')['total_vaccinations'].max().reset_index()
total
```

```python
from matplotlib.pyplot import figure 
figure(figsize=(50, 8), dpi=150)
total = h.groupby('country')['total_vaccinations'].max().reset_index()
plt.bar(total['country'],total['total_vaccinations'])
plt.xlabel("Country")
plt.ylabel("Total vaccinations")
plt.legend(["Totalvaccination"])
plt.xticks(rotation=90)
plt.show()
```
